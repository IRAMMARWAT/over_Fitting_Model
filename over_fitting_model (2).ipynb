```python
# Ensure a protobuf version compatible with TensorFlow to avoid:
# AttributeError: 'google._upb._message.FieldDescriptor' object has no attribute 'is_repeated'
# Install before importing tensorflow
%pip install --quiet protobuf==3.20.3

# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Load and preprocess the MNIST dataset
print("Loading MNIST dataset...")
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values to range [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")

# Define a function to create an overfit-prone model
def create_overfit_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(512, activation='relu'),
        Dense(512, activation='relu'),
        Dense(512, activation='relu'),
        Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

    return model

# Define a function to create an improved model
def create_improved_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

    return model

# Create both models
print("Creating models...")
overfit_model = create_overfit_model()
improved_model = create_improved_model()

print("Overfit Model Summary:")
overfit_model.summary()

print("\nImproved Model Summary:")
improved_model.summary()

# Set up Early Stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# Train the overfit-prone model
print("\nTraining overfit-prone model...")
history_overfit = overfit_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=128,
    validation_data=(X_test, y_test),
    verbose=1
)

# Train the improved model with early stopping
print("\nTraining improved model with early stopping...")
history_improved = improved_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=128,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping],
    verbose=1
)

# Evaluate both models
print("\nEvaluating models...")
overfit_loss, overfit_accuracy = overfit_model.evaluate(X_test, y_test, verbose=0)
improved_loss, improved_accuracy = improved_model.evaluate(X_test, y_test, verbose=0)

print(f"Overfit Model - Test Loss: {overfit_loss:.4f}, Test Accuracy: {overfit_accuracy:.4f}")
print(f"Improved Model - Test Loss: {improved_loss:.4f}, Test Accuracy: {improved_accuracy:.4f}")

# Plotting the results
plt.figure(figsize=(15, 5))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history_overfit.history['accuracy'], label='Overfit Train Accuracy')
plt.plot(history_overfit.history['val_accuracy'], label='Overfit Val Accuracy')
plt.plot(history_improved.history['accuracy'], label='Improved Train Accuracy')
plt.plot(history_improved.history['val_accuracy'], label='Improved Val Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_overfit.history['loss'], label='Overfit Train Loss')
plt.plot(history_overfit.history['val_loss'], label='Overfit Val Loss')
plt.plot(history_improved.history['loss'], label='Improved Train Loss')
plt.plot(history_improved.history['val_loss'], label='Improved Val Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()

# Additional analysis: Check for overfitting indicators
def analyze_overfitting(history, model_name):
    train_final_acc = history.history['accuracy'][-1]
    val_final_acc = history.history['val_accuracy'][-1]
    gap = train_final_acc - val_final_acc

    print(f"\n{model_name} Overfitting Analysis:")
    print(f"Final Training Accuracy: {train_final_acc:.4f}")
    print(f"Final Validation Accuracy: {val_final_acc:.4f}")
    print(f"Accuracy Gap: {gap:.4f}")

    if gap > 0.1:
        print("⚠️  Strong indication of overfitting!")
    elif gap > 0.05:
        print("⚠️  Moderate indication of overfitting")
    else:
        print("✅ Good generalization")

# Analyze both models
analyze_overfitting(history_overfit, "Overfit-Prone Model")
analyze_overfitting(history_improved, "Improved Model")

print("\nProject completed successfully!")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score

print("\n" + "="*50)
print("MACHINE LEARNING MODELS")
print("="*50)

def train_ml_models():
    """Train various Machine Learning models with overfitting analysis"""

    # Split training data further for validation in ML
    X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
        X_train_ml_scaled, y_train, test_size=0.2, random_state=42, stratify=y_train
    )

    # Define models - some prone to overfitting
    models = {
        'Overfit_RF': RandomForestClassifier(n_estimators=200, max_depth=None, min_samples_split=2, random_state=42),
        'Regularized_RF': RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=10, random_state=42),
        'Overfit_SVM': SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42),
        'Regularized_SVM': SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42),
        'Logistic_Regression': LogisticRegression(max_iter=1000, random_state=42)
    }

    results = {}

    for name, model in models.items():
        print(f"\nTraining {name}...")

        # Train model
        model.fit(X_train_split, y_train_split)

        # Predictions
        train_pred = model.predict(X_train_split)
        val_pred = model.predict(X_val_split)
        test_pred = model.predict(X_test_ml_scaled)

        # Calculate accuracies
        train_acc = accuracy_score(y_train_split, train_pred)
        val_acc = accuracy_score(y_val_split, val_pred)
        test_acc = accuracy_score(y_test, test_pred)

        # Cross-validation for better generalization estimate
        cv_scores = cross_val_score(model, X_train_ml_scaled, y_train, cv=3, scoring='accuracy')

        results[name] = {
            'train_accuracy': train_acc,
            'val_accuracy': val_acc,
            'test_accuracy': test_acc,
            'cv_mean': cv_scores.mean(),
            'cv_std': cv_scores.std(),
            'overfit_gap': train_acc - val_acc
        }

        print(f"  Train Accuracy: {train_acc:.4f}")
        print(f"  Val Accuracy: {val_acc:.4f}")
        print(f"  Test Accuracy: {test_acc:.4f}")
        print(f"  CV Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
        print(f"  Overfit Gap: {train_acc - val_acc:.4f}")

    return results

# Train ML models
ml_results = train_ml_models()
```

```python
# Load and prepare data if not already done (ensures X_train, y_train, etc. are available)
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import StandardScaler

# Load MNIST dataset if not already loaded
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Reshape for Machine Learning (flatten images)
X_train_ml = X_train.reshape(X_train.shape[0], -1)
X_test_ml = X_test.reshape(X_test.shape[0], -1)

# Normalize for ML
scaler = StandardScaler()
X_train_ml_scaled = scaler.fit_transform(X_train_ml.astype('float32'))
X_test_ml_scaled = scaler.transform(X_test_ml.astype('float32'))

print(f"ML Data shape: {X_train_ml_scaled.shape}")

# Train ML models
ml_results = train_ml_models()
```

```python
print("\n" + "="*50)
print("DEEP LEARNING MODELS")
print("="*50)

# Deep Learning imports (added for robustness)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau # Moved from c84386b9

def create_overfit_dl_model():
    """Create a deep learning model prone to overfitting"""
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(1024, activation='relu'),
        Dense(1024, activation='relu'),
        Dense(512, activation='relu'),
        Dense(512, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def create_regularized_dl_model():
    """Create a regularized deep learning model"""
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),
        Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def train_dl_models():
    """Train deep learning models with callbacks"""
    # Callbacks for regularization
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001, verbose=0)

    # Overfit model (no regularization)
    print("Training Overfit DL Model...")
    overfit_model = create_overfit_dl_model()
    history_overfit = overfit_model.fit(
        X_train_dl, y_train,
        epochs=50,
        batch_size=256,
        validation_data=(X_test_dl, y_test),
        verbose=1
    )

    # Regularized model
    print("\nTraining Regularized DL Model...")
    regularized_model = create_regularized_dl_model()
    history_regularized = regularized_model.fit(
        X_train_dl, y_train,
        epochs=50,
        batch_size=256,
        validation_data=(X_test_dl, y_test),
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )
    # Evaluate models
    overfit_loss, overfit_acc = overfit_model.evaluate(X_test_dl, y_test, verbose=0)
    regularized_loss, regularized_acc = regularized_model.evaluate(X_test_dl, y_test, verbose=0)

    dl_results = {
        'DL_Overfit': {
            'history': history_overfit,
            'test_accuracy': overfit_acc,
            'test_loss': overfit_loss
        },
        'DL_Regularized': {
            'history': history_regularized,
            'test_accuracy': regularized_acc,
            'test_loss': regularized_loss
        }
    }

    print(f"\nDL Overfit Model - Test Accuracy: {overfit_acc:.4f}")
    print(f"DL Regularized Model - Test Accuracy: {regularized_acc:.4f}")

    return dl_results
```

```python
# Ensure data for Deep Learning models is available
from tensorflow.keras.datasets import mnist # Ensure mnist is imported

# Load MNIST dataset if not already loaded
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values to range [0, 1] for Deep Learning
X_train_dl = X_train.astype('float32') / 255.0
X_test_dl = X_test.astype('float32') / 255.0

print(f"Ensured DL Data shape: {X_train_dl.shape}")

# Train DL models
dl_results = train_dl_models()
```
